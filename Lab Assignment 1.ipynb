```python
# Import an audio file using the librosa library
import numpy as np
import matplotlib.pylab as plt
import librosa
import librosa.display
```

```python
file = 'avm2.wav'
sig , sr = librosa.load(file,mono=True,sr=None)
```

```python
# Calculate the length in seconds
length_s = sig.shape[0] / sr
print(length_s)
```

```python
# Plot the time-domain representation of avm2.wav
plt.plot(sig)
```

```python
#Plot with time on the x-axis
time=np.arange(0,sig.shape[0])/sr
plt.plot(time,sig)
```

```python
# Plot with amplitude on the y-axis
excursion=max(abs(sig))
sigNew = sig / excursion
plt.plot(time,sigNew)
```

```python
# Plot with labels
time=np.arange(0,sig.shape[0])/sr
plt.plot(time,sigNew)
plt.xlabel('Time (Seconds)')
plt.ylabel('Amplitude')
```

```python
# Plot sampling rate of 44100
def plotAudio(filename,sampleRate):
    sig , sr = librosa.load(filename,mono=True,sr=sampleRate)
    plt.plot(time,sig)
    plt.xlabel('Time (Seconds)')
    plt.ylabel('Amplitude')
```

```python
plotAudio('avm2.wav',44100)
```
