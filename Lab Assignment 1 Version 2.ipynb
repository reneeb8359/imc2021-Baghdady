```python
# Import an audio file using the librosa library
import numpy as np
import matplotlib.pylab as plt
import librosa
import librosa.display
file = 'avm2.wav'
```

```python
# Plotting audio function 
def plotAudio(filename,sampleRate): 
    #Open the audio file with librosa.load
    sig , sr = librosa.load(filename,mono=True,sr=sampleRate)
    
   #Plot x-axis with time 
    time=np.arange(0,sig.shape[0])/sr
    
    # Plot  y-axis with amplitude 
    excursion=max(abs(sig))
    sigNew = sig / excursion
    
    # Plot x and y axis with variables
    time=np.arange(0,sig.shape[0])/sr
    plt.plot(time,sigNew)
    plt.xlabel('Time (Seconds)')
    plt.ylabel('Amplitude')
```

```python
plotAudio('avm2.wav',44100)
```

```python
plotAudio('avm2.wav',441)
```
